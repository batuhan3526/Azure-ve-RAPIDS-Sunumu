## Bulut Bilişim Nedir?

Son birkaç yıl, yeni teknolojilerin ortaya çıkması açısından devrim niteliğinde oldu. Büyük veri ve bulut bilişim gibi kavramlar hem kişisel hem de profesyonel kullanımda giderek yaygınlaştı. Bununla birlikte, birçok insan için bu terimler hala büyük bir bilinmez. 

Bulut bilişim; sunucu, depolama, veri tabanı, ağ, yazılım vb. gibi bilgi işlem hizmetlerinin internet üzerinden sağlanmasıdır. İngilizce karşılığı "cloud computing" olan bulut teknolojisi, sanal sunucularda depolanan verilerinize istediğiniz her yerden ve tüm cihazlardan kolaylıkla erişmenizi mümkün kılar.

ABD Ulusal Standartlar ve Teknoloji Enstitüsü bulut bilişimi, düşük yönetim çabası veya servis sağlayıcı etkileşimi ile, hızlı alınıp salıverilebilen ayarlanabilir bilişim kaynaklarının paylaşılır havuzuna, istendiğinde ve uygun bir şekilde ağ erişimi sağlayan bir modeldir, şeklinde tanımlamıştır.

Bulut teknolojisi yazılım, donanım, personel ve eğitime çok büyük yatırımlar yaparak bunları şirket içinde departmanlaştırmaktan ziyade yazılım ve donanım tesislerindeki kurulu sistemlere internet üzerinden düşük ücretlerle bağlanarak kullanma temeline dayanmaktadır.

e-Posta göndermek, internet üzerinden video veya film izlemek, müzik dinlemek, oyun oynamak, fotoğraf, belge saklamak için çevrimiçi bir hizmet kullanıyorsanız tüm bunların alt yapısı bulut teknolojisi tarafından sağlanmaktadır.

## BULUT BİLİŞİMİN TARİHÇESİ NEDİR?

Anlaşılması adına şöyle bir örnek verilebilir. Edison zamanında önemli bir sorunu gördü. Eğer ki şirket kurmak ve büyümek istiyorsanız ne iş yaparsanız yapın, kendi elektriğinizi üretmeniz lazımdı. Çünkü fosil yakıtlar geride kalmaya başlamıştı.
Ama her şirketin kendi elektriğini üretmesi oldukça masraf ve yüksek yatırım demekti. Edison ise onlara "Ben sizin yerinize elektriği üretirim. Siz hiç bu alana yatırım yapmayın benden alın elektriğinizi ve kullandığınız kadar ödeyin" dedi kabaca. Daha sonra ise birden patladı elbette. Bulut sistemleri de tamamen aynı mantık üzerine kuruludur. 

Dünyanın en büyük bulut altyapı hizmetleri sağlayıcılarından Amazon, veri merkezlerini revize ederek ilk bulut teknoloji hizmeti olan Amazon S3'ü, 2006'da hizmete soktu. Ardından 2008 senesinde Danışmanlık ve Araştırma Şirketi Gartner, bulut bilişimin BT hizmetleri sektöründe kullanıcılar ve tedarikçiler arasındaki ilişkiyi farklı boyuta taşıyabilecek güce sahip olduğunu ortaya koydu. Bulut teknolojisi bu tarihten itibaren tüm dünyada kullanılmaya başlandı. Özellikle son yıllarda ülkemizde de benimsenerek kullanılan bulut bilişim, oldukça hızlı bir biçimde büyümeye devam etmektedir.

# Bulut Bilişim Nasıl Çalışır?

Bulut bilişimin nasıl çalıştığını anlamak söz konusu olduğunda, bulut bilişimi iki arayüz açısından incelemek faydalı olabilir. Bunlar, bilgi işlem sisteminin ön ucu ( frontend) ve arka ucudur (backend). Sistemin ön ucu, kullanıcıların internet web tarayıcıları veya bulut bilişim uygulamaları aracılığıyla bulutta depolanan verilere erişmesine olanak tanır.

Ancak bulut bilişimin ana bileşeni arka uçla ilgilidir. Bu, verilerin ve bilgilerin depolanmasına izin veren işlevsellikten sorumlu olan sistemin yönüdür.

Arka uç; merkezi sunucular, veri tabanları ve bilgisayarlar dahil olmak üzere sunuculardan oluşur.

Merkezi sunucu, bulut bilişim ile bağlanan cihazlar arasında sorunsuz bağlantılar sağlayan bir yazılım türü olan “ara yazılım” kullanır. Bulut sağlayıcılar, güvenlik ihlalleri ve veri kaybı riskini azaltmak için genellikle birden çok veri kopyasını saklar.

İşletmeler, kendi BT altyapılarını veya veri merkezlerini satın alıp çalıştırmak yerine, bulut hizmet sağlayıcılarından uygulamalar ve depolama dahil olmak üzere kaynaklara erişim kiralayabilir veya ihtiyaç duydukları kadar hizmet alabilir.

## BULUT BİLİŞİMİN ÖZELLİKLERİ NELERDİR?

Bulut bilişimin en önemli özelliği, donanım ve yazılımın sanallaştırılması ve hizmet odaklı olmasıdır. Bu nedenle her geçen gün daha fazla şirket dijital dönüşüme ayak uydurabilmek için bulut teknolojisini tercih ediyor.

Şirket içinde kurulan sunucuların ve veri depolama uygulamalarının işletmelere getirdiği donanım yenileme, yazılım güncelleme, bakım, depolama gibi iş yükü ve maliyetlerini ortadan kaldıran bulut bilişim teknolojisi, big data çağında veri saklama ve işlemede en doğru çözüm.


Bulut teknolojisiyle birlikte boyutlandırma, yapılandırma ya da yedekleme gibi manuel işlemler otomatik hale gelerek maliyet ve zaman tasarrufu sağlıyor.
Kurumların çalışabilirlik sürelerini artırmalarında, buluttan ve büyük veriden toplanan datanın güvenle depolanabilmesinde, iş sürekliliği ve felaket önleme senaryolarının oluşturulmasında sanallaştırma teknolojileri önemli bir rol oynuyor.
Sanallaştırma teknolojileri varolon kaynaklar üzerinde ayrıntılı bir kontrol mekanizması oluşturulmasına imkan tanır. Özellikle performans problemlerinden kaynaklanan sorunlara çok daha hızlı müdahale etmek mümkün olur.


Bulut bilişim, bilgi teknoloji operasyonlarına gerekli kaynağı sağlayamayan işletmelerin veri merkezi, donanım, yazılım, ve bilgi teknolojisi uzmanları gibi yüksek bütçe gerektiren yatırımları yapmaktan ziyade bulut üzerinden kiralama modeli ile kullandıkları kadar ödeyebilecekleri esnek bir sistemdir. Böylece BT yatırımları kontrol altına alınır ve temel iş koluna yoğunlaşarak rekabette avantaj sağlamak mümkün olur.

Kullandığın kadar öde yöntemiyle kullanmayacağınız özelliklerde bir donanım sahibi olmak yerine, ihtiyacınız olduğu kadarıyla donanımı düşürebilir veya yükseltebilirsiniz.
Bulut bilişimde donanım ve yazılım uygulamaları daima en son sürümüne güncellenir. Güncelleme işlemlerinde temel sistem görüntüsü üzerinden yapılan değişiklikler tüm cihazlara otomatik olarak yansıtılır.
Bulut bilişim altyapılarında yedekli çalışan sistemler sayesinde veri kaybı ve işleyişte aksaklıkların olma olasılığı minimum düzeydedir.

Bulut depolama güvenlidir. İsteğe bağlı olarak ölçeklendirme yapıldığından ihtiyaç olan depolama alanına gerektiğinde sahip olunabilir.
Dijital bir ağ aracılığıyla çoklu sunucu bağlantısı sayesinde en düşük kapasiteli cihazla dahi istenilen yerden istenilen zamanda her tür bilgiye erişim sağlanır.
Bulut bilişim, kişisel bilgisayarların yükünü azaltır ve çok sayıda yazılım, bulut sunucusu tarafından sağlanır.
Nihai kullanıcıya eğitici ve yönlendiricilerle kolay kullanım özelliğine sahiptir.

# Bulut bilişimi kimler kullanıyor?

Her türden, boyuttan ve sektörden kurum, bulutu veri yedekleme, olağanüstü durum kurtarma, e-posta, sanal masaüstleri, yazılım geliştirme, test etme, büyük veri analizi ve müşterilere dönük web uygulamaları gibi çok çeşitli alanlarda kullanmaktadır.

* Örneğin, sağlık hizmetleri şirketleri, bulutu hastalar için daha kişiselleştirilmiş tedaviler geliştirmek için kullanır. Finansal hizmet şirketleri, bulutu gerçek zamanlı dolandırıcılık algılamayı ve önlemeyi desteklemek için kullanır. Bilgisayar oyunu geliştiricileri, bulutu dünya çapında milyonlarca oyuncuya çevrimiçi oyunlar sunmak için kullanır.


Bireysel açıdan ise, içerik üreticiler, Bilimsel hesaplamalar, AI, Video işleme, Mimarlık ve mühendislik çözümleri ile ilgilenenler kullanmaktadır. Sayıları az da olsa, Bulut üzerinden tasarım, mining yapan ve oyun oynayan insanlara denk gelebilirsiniz.

## BULUT BİLİŞİMİN AVANTAJLARI NELERDİR?

* Bulut bilişim (cloud computing), yazılım, donanım ve kurum içi veri merkezleri oluşturma gibi yatırım maliyetlerini ortadan kaldırır.

* Bilgi teknolojileri kaynağına gereksinim duyulan zamanda ve istenen coğrafyada erişebilme imkanına sahiptir.
* Bulut bilişim uygulamasının kurulması için bir BT uzmanına ihtiyacınız olmaz. Ayrıca güncellemeler otomatik olarak yapılır.
* Bulut teknolojisiyle gün geçtikçe çoğalan verilerin takibi, saklanması ve kullanıcı yetkilendirilmesi gibi alt yapı sorunları ortadan kalkar.
* Bilgi işlem kaynaklarına dakikalar hatta saniyeler içinde erişim sağlanabilir, dolayısıyla iş süreçleri hız ve çeviklik kazanır.
* Sınırsız veriden faydalanma imkanı sağlar.
* Bulut bilişim sayesinde yapay zeka (AI), Nesnelerin İnterneti (IoT), blokzinciri (Blockchain) ve sohbet robotları gibi yenilikçi teknolojilerin kullanımı dahil olmak üzere yeni uygulama yazılımları ve servislerine sahip olunabilir.
* Kurum içi veri merkezleri donanım, yazılım vs. gibi bilgi teknolojisi yönetim görevlerinden oluşan pek çok iş yükü gerektirir. Oysa bulut bilişim, bu görevlere duyulan gereksinimi ortadan kaldırır.
* Bulut teknoloji hizmetleri, dünya ölçeğindeki en güvenli veri merkezleri sunucularında çalışır. Bu veri merkezleri belli periyotlarda, hızlı ve verimli bilgi işlem donanımlarının en güncel sürümüne yükseltilir.
* Bulut bilişim hizmet sağlayıcılarının serverları, yazılım ve donanım güvenlik önlemlerini 7/24 sağladıkları için ana bilgisayardan çok daha güvenlidir.
* İsteğe bağlı olarak ölçeklendirme yapılabildiğinden, ihtiyacınız olan depolama alanına gereksinim duyduğunuz zaman sahip olursunuz.
* Yüksek bütçeli güvenlik ve depolama yazılımlarına servet harcamak yerine bulut bilişimi kullandığınız sürece ödeme yaparsınız.

Başlıklar altında toplamak ister isek:


### Maliyet tasarrufu: 

Bulut bilgi işlem; donanım ve yazılım satın alma, şirket içi veri merkezleri kurma ve çalıştırma gibi yatırım giderlerini (sunucu rafları, güç ve soğutma için devamlı elektrik ve altyapının yönetilmesi için çalıştırılan BT uzmanları) ortadan kaldırır. Bu da masraflar açısından oldukça önemli bir avantajdır.


### Performans:

En büyük bulut bilgi işlem hizmetleri, dünya çapındaki güvenli veri merkezleri ağında çalışır. Bu veri merkezleri düzenli olarak, hızlı ve verimli bilgi işlem donanımlarının en son sürümüne yükseltilir. Bu, tek bir kurumsal veri merkezine göre, uygulamalarda daha düşük ağ gecikme süresi ve çok daha ekonomik ölçeklendirme gibi birçok avantaj sunar.

### Hız:

Birçok bulut bilgi işlem hizmetinin self servis ve isteğe bağlı olarak sunulmasıyla çok büyük çaptaki bilgi işlem kaynakları bile yalnızca birkaç tıklamayla dakikalar içinde sağlanabilir, bu sayede işletmeler daha fazla esneklik kazanır ve kapasite planlama baskısından kurtulur.

### Verimlilik:

Şirket içi veri merkezleri genellikle çok sayıda “raflama ve yığınlama” işlemi gerektirir. Bunlar donanım ayarlama, yazılımlara düzeltme eki uygulama ve zaman alan diğer BT yönetim görevlerinden oluşan işlemlerdir. Bulut bilgi işlem, bu görevlerin çoğuna yönelik ihtiyacı ortadan kaldırarak BT ekiplerinin zamanı daha önemli iş hedefleri için kullanmasını sağlar.

### Güvenilirlik:

Bulut bilgi işlem, verileri bulut sağlayıcısının ağındaki birden fazla yedekli konumda yansıtabilme olanağı sayesinde veri yedeklemeyi, olağanüstü durum kurtarmayı ve iş sürekliliğini kolaylaştırır ve daha ekonomik hale getirir.


### Veri güvenliği:

Devasa dijital veri tabanları temelinde çalışan bir iş dünyasında, işletmelerin sahip olduğu en büyük endişelerden biri verilerinin güvenliğidir. 


### Kontrol:

Bulut bilişim, modern işletmeler için özelleştirilebilir çözümler sunarken, işletmelerin hassas verileri üzerinde de tam kontrol sahibi olmalarını sağlar. Ekip liderleri, yalnızca birkaç tıklamayla erişim haklarını ayarlayabilir ve hiçbir bilginin yanlış ellere geçmemesini sağlayabilir. Aynı zamanda belgeler, iş akışlarını kolaylaştıran ve işbirliği çabalarını geliştiren doğru ekip üyelerine kolayca atanabilir. Birden fazla kişi aynı proje üzerinde gerçek zamanlı olarak çalışabilir ve güncellenmiş kopyaları e-posta yoluyla göndermek zorunda kalmaz.

### Rekabet avantajı:

Tamamen veya kısmen yeni bir BT altyapısına geçiş, bir ilk yatırımı ve bir inanç sıçramasını gerektirse de, bulutu kullanmak uzun vadede işinizi olumlu yönde etkileyecek ve daha fazla değer üretecektir.


## Bulutun Bazı Temel Terimleri:

### SLA ( High Availability):

Sistem kesintiye uğrasa bile ne kadar inaktif kalacağını belirtir. Bu hizmeti sunan şirketler %99 veya %99.9 gibi sayılar ile pazarlama yaparlar.

### Scalability ( Ölçeklenebilirlik): 

İhtiyaca göre kullanılan hizmeti arttırıp azaltmaktır ve manueldir. Örnek olarak, Bir üniversitenin Vize dönemi yoğunluk olacağını bildiklerinden vize zamanı manuel bir şekilde alınan hizmeti büyütürler(!)

### Elasticity ( Esneklik):

Scalability ile aynı işlevdedir. Tek farkı bu işlemin otomatik olmasıdır. Böylece beklenmedik durumlara karşı hazırlıklı olunur. Otomatik olma işlevi şöyle olabilir : " CPU kullanımı X saniye/dakika boyunca %90'ı geçerse, Y kadar Core takviye et".
Veya tam tersi durumda beklenenden az talep, daha az kullanım görülürse fazla olan hizmeti iptal edebilir ve böylece maddi zararı kısmış oluruz.


### Agility (Çeviklik):

Bulut, daha hızlı yenilik yapabilmeniz ve hayal edebildiğiniz neredeyse her şeyi yapabilmeniz için çok çeşitli teknolojilere kolayca erişmenizi sağlar. İhtiyaç duyduğunuzda işlem, depolama ve veritabanı gibi altyapı hizmetlerinden Nesnelerin İnterneti, makine öğrenimi, data lake'ler, analiz ve daha fazlasına kadar kaynaklarınızı hızla kullanmaya başlayabilirsiniz.

Teknoloji hizmetlerini dakikalar içinde dağıtabilir ve fikir aşamasından uygulama aşamasına, öncekine göre birkaç kat daha hızlı bir şekilde geçebilirsiniz. Bu da size denemeler yapma, yeni fikirleri farklı müşteri deneyimlerinde test etme ve işletmenizi dönüştürme özgürlüğü verir.

Çeviklik önemlidir çünkü Yeni start-upların en büyük sorunlarından birisi Server maliyetleridir. Server'ların hazırlanması ve size ulaşması genelde 6 ay civarında olur. Sadece soğutma maliyetleri bile onbinlerce dolarları bulabilir. Bakım vs. gibi maliyetleri siz düşünün. Bu süre zarfında rakip şirketlere koz vermek istemiyor iseniz bulut kullanmak zorundasınız. 

### Latency ( Gecikme) :

Bildiğiniz üzere Server ile aranızdaki mesafe arttıkça gecikme artacaktır. Bu nedenle büyük devler her yere Server açmaktadırlar.

### Security (Güvenlik) :

Çoğu bulut sağlayıcısı, güvenlik konusundaki genel duruşunuzu güçlendiren çeşitli ilkeler, teknolojiler ve denetimler sunarak verilerinizi, uygulamalarınızı ve altyapınızı olası tehditlere karşı korumanıza yardımcı olur.

### Disaster Recovery (Olağanüstü durum kurtarma) :

Veriler yerel sabit disk ve sunucularda depolandığında, veri kaybı bir şirket için büyük bir tehdit oluşturur. Bulut altyapısı ile veriler yalnızca 7/24 mevcut olmakla kalmaz, aynı zamanda otomatik olarak yedeklenir.

### Fault Tolerance ( Hata Toleransı) :

Server'lar da elektronik bileşenlerden oluştuğu için zamanla bozulabilirler. Bu nedenle önemli verileriniz bu riski almadan yedeklenir. Çoğu şirket için olmazsa olmazdır.

### Predictive Cost Considerations ( Tahmini Maliyet Konusu) :

Kullanmış olduğunuz hizmetin tahmini faturasını gördüğünüz sekmedir.

## BULUT BİLİŞİM HİZMET MODELLERİ NELERDİR?

Bilgi sistem türleri olarak da bilinen bulut bilişim hizmeti, 3 kategoriden oluşmaktadır. Iaas, Paas ve SaaS'ın ne olduğunu ve aralarındaki farkları bilmek ihtiyaçlarınıza uygun hizmet modelini belirlemenize ve iş hedeflerinizi gerçekleştirmenizi yardımcı olacaktır.





***  Böyle tam anlaşılmıyorsa Restoran üzerinden gidelim.***

### IaaS (Infrastructure as a Service) Sunucu Altyapı Hizmeti Nedir?

IaaS, bulut bilişim teknolojisinin sağladığı en temel ve en esnek bulut bilgi işlem modelidir. Altyapı hizmeti ile sanal sunucu oluşturularak bulut bilişim "cloud computing" hizmeti sağlanmaktadır. Depolama, network ve sanallaştırma gibi hizmetler sunan IaaS, kullandıkça öde sistemine dayalıdır.

Amazon Web Services (AWS), Google Compute Engine (GCE) ve Microsoft Azure bulut (cloud) tabanlı altyapı hizmetlerindendir.

#### IaaS Avantajları Nelerdir?

* IaaS, oldukça esnek ve ölçeklenebilir çözümler sunar ayrıca düşük maliyetlidir.
* IaaS sayesinde fiziksel sunucular ve veri merkezi altyapısı oluşturmanıza gerek kalmaz.
* IaaS kullanılarak çalıştırılan web siteleri, geleneksel web barındırmaya oranla daha az maliyetlidir.
* Test etme ve geliştirme ortamları çok daha hızlı şekilde oluşturulabilir. Bu sayede yeni uygulamalar pazara çok daha hızlı şekilde sürülebilir.
* IaaS depolama, network ve işletim sistemlerinden kiralama usulü ile yararlanılır.
* IaaS sisteminde sadece kullandığınız kaynaklar için ödeme yaparsınız.

### PaaS (Platform as a Service) Platform Hizmeti Nedir?

IaaS'ın daha fazla gelişmiş ve özellikli hedeflere yönelik olan platformudur. PaaS, uygulama geliştirenlere, projelerini geliştirebilmeleri için internet üzerinden donanım ve yazılım araçları sağlayan bulut bilişim hizmet modelidir.

PaaS, altyapının (sunucular, depolama ve network) yanı sıra ara yazılım, geliştirme araçları, iş zekası hizmetleri, veritabanı yönetimi, sistem yönetimi ve güvenliğe yönelik çözümleri de kapsamaktadır. PaaS kullanıcılara uygulama geliştirecekleri, uygulamaları yönetecekleri ve sunacakları bir bulut ortamı sağlar. Programlama dili, veri tabanı ve işletim sistemi vs. PaaS sistemine en iyi örnektir.

PaaS'ta depolama ve kurulum alanına ilaveten kullanıcıya bir bilişim platformu ve özgül çözüm yazılımları da sunulur. Çoğunlukla belli sistemler için bulut çözümleri geliştiren, test eden ve kurduktan sonra üçüncü kişilere kullandıran kurumlar tarafından tercih edilir. Bu nedenle PaaS hizmet sağlayıcı, barındırmayla birlikte uygulama geliştiriciye yalıtılmış bir kurulum, test ortamı, işletim sistemi, bakım ve güncelleme desteği sağlar.

PaaS'ın bilinen örnekleri arasında Microsoft Azure ve Google App Engine sayılabilir.

#### PaaS Avantajları Nelerdir?

* PaaS, IaaS'nin sunduğu avantajlarla birlikte ara yazılım, geliştirme araçları ve diğer iş araçları gibi ek özellikler de sunar.
* PaaS, uygulama geliştiricilerin, bilgi teknoloji yöneticilerinin ve nihai kullanıcıların mobil ve web uygulama yazılımları oluşturmasına, entegre etmesine, taşımasına, dağıtmasına, güvenliğini sağlamasına ve yönetmesine imkan tanır.
* PaaS, uygulama geliştiricilerin yazılımlarında yapay zeka (AI), nesnelerin interneti (IoT), blokzinciri, sohbet robotları ve inovatif teknolojileri de kapsayan yeni özellikler oluşturmasına imkan tanıyan kullanıma hazır programlama bileşenleri de sunar.
* PaaS geliştirme araçları sayesinde yeni uygulamaların kodlama sürecini kısaltır.
* PaaS, web uygulaması oluşturma, test etme, dağıtma, yönetme ve güncelleştirme için ihtiyacınız olan tüm özellikleri sağlar.
* PaaS, çoğul kullanıcılar tarafından erişilebilir ayrıca işin büyüklüğüne göre ölçeklendirilebilirsiniz.
* Güvenlik, işletim sistemleri, sunucu yazılımları ve yedeklemeler hizmet sağlayıcılar tarafından yönetilir. Bu durum iş yükünü azaltır.
* Ekip üyeleri farklı coğrafyada olsalar dahi projeler üzerinde birlikte çalışabilir.
* PaaS sayesinde bilgi teknoloji operasyonları standart ve daha kolay hale gelir.
* Operasyonel, güvenlik ve yönetim riskleri azalır.
* Kullanıma hazır çözümlerle iş inovasyonu hızlanır.
* PaaS, geliştiricilerin yazılım güncellemeleri gibi iş yükü yerine yaratıcılığa konsantre olmalarına imkan tanır.
* Kapsamlı sistem yönetimi bilgisi olmadan kolaylıkla çalıştırılabilir.
* PaaS, uygulama geliştirmenin en uygun maliyetli yoludur. Kullandıkça öde esasına dayalıdır.

### SaaS (Software as a Service) Yazılım Hizmeti Nedir?

SaaS hizmet sağlayıcılar, nihai kullanıcıya web tabanlı uygulamalar sunar. Ayrıca bu hizmetlerin güncel olmasını ve güvenliğini sağlar. Altyapı, yazılım ve tüm data SaaS tedarikçisinin veri merkezinde tutulur. Tedarikçi, donanım ve yazılımı yöneterek uygulamaların, verilerin kullanılabilirliğinden ve güvenliğinden sorumludur. SaaS, bir bulut hizmet sağlayıcısından kullandıkça öde esasına dayalı olarak satın alabileceğiniz kusursuz bir yazılım sistemidir.

İnternet tabanlı bir e-Posta hizmeti kullanıyorsanız, bir SaaS yazılım hizmeti kullanıyorsunuz demektir. Bu tarz hizmetlere, çoğunlukla bir web tarayıcısı aracılığıyla, hesabınıza kullanıcı adı ve parolanızla giriş yaptıktan sonra ulaşırsınız. e-Posta yazılımı, hizmet sağlayıcısının ağında bulunur ve iletileriniz de burada depolanır. e-Postalarınıza bilgisayar ya da mobil cihazlardaki web tarayıcısından erişim sağlarsınız. Bu örnek şahsi kullanıma yöneliktir ve genellikle ücretsizdir.

Kurumsal kullanım mevzu bahis olduğu zaman, e-Posta, işbirliği, müşteri ilişkileri yönetimi (CRM), kurumsal kaynak planlama (ERP) ve belge yönetimi vs. gibi gelişmiş kurumsal uygulamaları kiralamanız gerekmektedir. Söz konusu uygulamalar için abone olarak ya da kullanım seviyesine göre ödeme yapmanız gerekir.

SaaS'ın en bilindik örnekleri Microsoft Office 365, Google Apps, Dropbox'dır.

#### SaaS Avantajları Nelerdir?
* Kullanıcılar SaaS uygulamalarının çoğunu bir program indirip yüklemeye ihtiyaç duymadan direkt olarak çalıştırabilir. Dolayısıyla kullanıcılarınız için özel bir program satın alarak kurmanıza gerek yoktur.
* Uygulamalar, web ya da bir API üzerinden erişilebilen, bir bulut ağında bulunduğu için kullanıcıların uygulamaları yerel cihazlarına kurmaları gerekmez.
* Veriler bulutta saklanır ve son derece güvenlidir.
* Verilerin bulutta depolanması sayesinde kullanıcılar, bilgilerine bilgisayar veya mobil aygıtlardan erişim sağlar. Uygulama verilerinin bulutta depolanması sayesinde, kullanıcı bilgisayarı veya mobil cihazı çökse dahi veriler güvenle saklanır.
* Kullanıcılar, uygulamaları kullanarak verileri depolayabilir, analiz edebilir ve projeler üzerinde işbirliği yapabilir.
* Kullanıcıların yazılımları kurmaları, yönetmeleri ya da yükseltmelerine gerek yoktur. Tüm bu işlemler SaaS sağlayıcıları tarafından gerçekleştirilir.
* Kaynakların kullanımı hizmet gereksinimlerine göre ölçeklendirilebilir.
* Uygulamalara dünyanın hangi coğrafyasında olursanız olun, internet olan her yerden kolaylıkla erişim sağlanabilir.
* SaaS hizmet sağlayıcıları, kullanıcılara yazılım ve uygulamaları abonelik modeli üzerinden sunar.
* SaaS hizmetinde kullandığın kadar öde yöntemi geçerlidir. Bu nedenle kullanım seviyesine bağlı olarak otomatik olarak artırıp azalttığı için maddi açıdan tasarruf etmenizi de sağlar.

## BULUT BİLİŞİM DAĞITIM MODELLERİ NELERDİR?

### Public Cloud (Genel Bulut):
Hizmet sağlayıcının kaynaklarını internet üzerinden kullanıcılara açmasıdır. Genel bulut, komple buluta dağıtılmış ve tüm kısımları bulutta çalışan bir uygulamadır. Ekseriyetle web tabanlı e-Posta, ofis uygulamaları, depolama, test etme ve geliştirme maksadıyla kullanılmaktadır. Google ve Microsoft genel bulut sağlayıcılarındandır.

### Private Cloud (Özel Bulut):
Bulut teknolojisi kaynaklarının (donanım, depolama, uygulama) tamamının sadece bir şirket tarafından yönetilmesi ve kullanılmasıdır. Şirket içi bulut da denilmektedir. Özel bulut, bulut teknolojisi ile sunulan pek çok avantajdan yoksundur. Çoğu zaman geleneksel Bilgi Teknolojisi altyapısıyla aynıdır ancak kaynak kullanımını artırmak maksadıyla uygulama yönetimi ve sanallaştırma teknolojilerinden yararlanılmaktadır.

Özel bulutta kaynaklar şirketin hissedarları, çalışanları ve müşterileri tarafından internet üzerinden veya şirket içi ağ üzerinden kullanılır.

Özel bulut, şirketin tesisinde veya bulut sağlayıcının veri merkezinde barındırılır. Özel bulutu genellikle veri gizliliğine önem veren büyük kuruluşlar tercih etmektedir.

### Hybrid Cloud (Melez Bulut):
Hem genel hem de özel bulutların bileşiminden ortaya çıkan bulut teknolojisidir. Hybrid, bulut tabanlı kaynaklarla bulutta olmayan mevcut altyapı ve uygulamalar arasında bağlantı kurmak için kullanılan bir metottur.

Hybrid bulut, bir şirketin kritik iş yükleri için şirket içi özel bulutu; arşivleme, test etme, geliştirme vs. iş süreçlerinde ise genel bulutun kullanılmasını sağlar.

# Bulut Bilişim Tecihleri

Bildiğiniz gibi Pazarda pek çok oyuncu bulunmakta. Amazon'dan AWS, Google'dan GCP, Microsoft'tan Azure en bilinenleridir. Bunlar haricinde IBM, Alibaba gibi şirketler de hizmet vermektedir.

AWS 2006, Azure 2010, GCP 2011 yılında açıldı. AWS şirketleri ve devletleri hedefler iken, GCP youtube ve diğer platformları desteklemeyi ve bireysel kullanıcıyı amaçladı. Azure ise 2 alanıda kapsayacak bir politika belirledi.


## Erişilebilirlik

* AWS'nin 66 erişilebilirlik bölgesi var ve 12'si daha yolda.
* Azure, dünya çapında 54 bölgeye sahiptir ve tüm dünyada 140 ülkede kullanılabilir.
* Google Cloud Platform, dünya çapında 20 bölgede kullanıma sunuldu ve 3 tane daha yolda.

## Karlılık

2021 için bildirilen üç aylık kazançlara göre, Microsoft'un Azure bulut gelirinin bir kez daha hem AWS hem de Google Cloud'un toplamından daha iyi performans gösterdiği gözlemlendi.



AWS, pazarın %32'sini kapsıyor, ardından %19 ile Azure ve %7 ile Google geliyor.


AWS'nin önde olmasının temel sebebi piyasaya ilk giren olduğu tahmin ediliyor. Ama Azure'nin hızlı yükselişi onu birkaç yıl içinde tahtından edebilir.


## Neler Sunuyorlar?

Hemen hemen hepsi aynı şeyleri sunuyor olsalar da, Hizmet sayısı olarak AWS öndedir ( Buna Rağmen Azure ile aralarında ki fark fazla değildir). Lakin GCP ve Azure hızla yeni Servisler eklemektedirler. 

Hizmetler aynı olsa da isimler değişmektedir. Örneğin Depolama için şöyledir

Object Storage - AWS Amazon Simple Storage Service (S3) - Azure Blob Storage - GCP Google Cloud Storage

AWS şirketlere yönelik çözümler ve analitik ile ön plana çıkarken, Azure Microsoft yazılımlarını desteklemeye ve Bilişsel Hizmetler vermeye odaklanmış durumda. Google ise bireysel kullanıcılara ve IoT'ye odaklanmış durumadadır.


## Artılar ve Eksiler

### AWS

AWS'nin yaklaşık yüzde 33'lük toplam pazar payını kapsayan bulut bilişim endüstrisindeki en büyük oyuncu olması hiç de akıllıca değil. Bu popülerliğin bariz nedenlerinden biri, AWS tarafından sunulan 200'den fazla yönetilen hizmet ve bunların kullanım kolaylığıdır.

Buna ek olarak, AWS'yi birinci sınıf bir pazar oyuncusu yapan başka avantajlar da vardır. Muazzam bir operasyon kapsamına ve dünya çapında kapsamlı bir veri merkezleri ağına sahiptir. Kullanıcılarına ölçeklenebilirlik ve bütünsel güvenlik sağlama kolaylığı ile AWS, en olgun ve kurumsal kullanıma hazır sağlayıcı haline geldi.

Bu avantajlara sahip olmanın yanı sıra, AWS'nin fiyatlandırma stratejisinde bir dezavantajı vardır. Kuruluşlar AWS'yi en uygun bulut hizmeti sağlayıcısı olarak görse de, genellikle fiyatlandırma stratejisi konusunda kafaları karışır. Birçok kuruluş, fiyatlarını sürekli olarak düşürdükten sonra bile, hizmet üzerinde yüksek hacimli iş yükleri çalıştırırken AWS'nin maliyet yapısını anlamakta ve bu maliyetleri etkin bir şekilde yönetmekte zorlanıyor. 

### Azure

Microsoft, Windows Server, Office, SQL Server, Sharepoint ve diğerleri gibi şirket içi hizmetlerini buluta alarak bulut pazarına girdi. Azure, kuruluşların çoğunluğu tarafından yaygın olarak kullanılan diğer uygulamalarla entegre olduğu için bu, Microsoft'un rakiplerini geride bırakmasına yardımcı olur. Ayrıca Micorosoft, müşterilerine hizmet sözleşmelerinde de önemli indirimler sağlıyor. AWS'ye kıyasla daha ucuz olması ve Microsoft müşterisi iseniz  extra indiminiz sayesinde ucuza hizmet alabiliyorsunuz. AWS daha çok kurumsalları hedef alıyor diye Azure'nin eli armut toplamadı. Fortune 500 şirketlerinin %95'inden fazlası tarafından kullanılan Microsoft Azure, kurumsal kullanıcılara hizmet sağlama konusunda kanıtlanmış bir sicile sahiptir. Kurumların Buluta entegrasyonu konusunda 

Microsoft'un yetersiz kaldığı alanlardan bazıları, platform için gereken bakım ve bazı hizmetlerin rakiplerine kıyasla sönük kalmasıdır. Azurenin  kuvvetli oldugu alanlar şirket içi yazılımlar, bilişsel hizmetler ve veri tabanlarıdır.

### GCP

Büyük veri, analitik ve makine öğrenimi gibi yüksek bilgi işlem tekliflerinde uzmanlaşan Google bulut, önemli ölçüde ölçeklendirme ve yük dengeleme özellikleri sunar. Bireysel kullanıcı daha çok hedefi olduğundan depolama alanında diğerlerine nazaran daha uygun oldugunu söyleyebiliriz.

Google Cloud'un belirli avantajları olsa da birkaç dezavantajı da vardır. Google'ın kurumsal müşterilerle geleneksel bir ilişkisi yoktur. Ancak, tekliflerini ve küresel veri merkezlerinin ayak izini hızla genişletiyor.


Özetlemek gerekirse AWS tam bir Goliath olmaya çalışırken, MS araçlarını Azure bulut ile tümleştirmek kolay olduğundan, MS araçlarını kullanan birkaç kuruluş için Azure bulutu kullanmak daha mantıklıdır. Ayrıca, Google Arama motorunun ve YouTube'un üzerinde çalıştığı altyapı için en iyi fiyatlandırma modelini sunan GCP, müşterilerin GCP'yi tercih etmesinin tek nedenidir.
