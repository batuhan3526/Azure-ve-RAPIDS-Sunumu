# RAPIDS Nedir, Yenir mi ?



RAPIDS, uçtan uca (end of end) veri bilimi ve analitik işlem hatlarını (pipeline) tamamen grafik işleme birimlerinde (GPU'lar) yürütmek için açık kaynaklı yazılım kitaplıkları paketidir. RAPIDS, daha üretken iş akışları oluşturmak için veri bilimi işlem hatlarını hızlandırır.RAPIDS, uçtan uca veri bilimi işlem hatları için büyük veri kümesi boyutlarında çoklu GPU dağıtımlarını da destekler.


RAPIDS MEDIUM : https://medium.com/rapids-ai


## RAPIDS Nasıl Çalışır?

RAPIDS, Apache Arrow sütunlu bellek platformuna dayalı bir GPU Veri Çerçevesi (DataFrame) kullanan makine öğrenimi topluluğundaki ortaklarla birlikte GPU Open Analytics Initiative (GoAI) tarafından 2017 yılında oluşturuldu. Amaç, GPU'larda uçtan uca veri bilimi ve analitik işlem hatlarını hızlandırmaktı.

RAPIDS, tüm veri bilimi ve analitik iş akışlarının daha hızlı çalışmasını sağlamak için GPU hızlandırmalı makine öğrenimini kullanır. GPU için optimize edilmiş bir çekirdek veri çerçevesi, veritabanları ve makine öğrenimi uygulamaları oluşturmaya yardımcı olur. RAPIDS, Python gibi basit şekilde tasarlanmıştır 

RAPIDS, makine öğrenimi algoritmalarıyla bütünleşen bir Veri Çerçevesi API'si içerir.

## RAPIDS, Veri Bilimi ve Analitik İşlem Hatlarını Nasıl İyileştirir?

RAPIDS, daha etkileşimli ve keşif amaçlı iş akışlarına izin vermek için veri yükleme, ETL, model eğitimi ve çıkarım dahil olmak üzere yapılabilecek hemen her şeyi GPU üzerinde yapılmasını sağlayarak veri bilimi işlem hattını hızlandırır.

* RAPIDS'in faydaları aşağıda ki gibi sıralanabilir:

1) Entegrasyon — Minimum kod değişikliği ile Python veri bilimi araç zincirini hızlandırın.

2) Ölçekleme — Çoklu GPU dağıtımları ve çok düğümlü kümeler dahil olmak üzere herhangi bir GPU'da sorunsuz ölçeklendirme.

3) Doğruluk — Makine öğrenimi modeli doğruluğunu artırmak için daha hızlı model dağıtımına ve yinelemelere olanak tanır.

4) Hız — Veri bilimi üretkenliğini artırmak için daha hızlı eğitim süresi.

5) Açık kaynak — NVIDIA tarafından desteklenen ve Apache Arrow üzerine kurulu özelleştirilebilir açık kaynaklı yazılımdır.

## RAPIDS GPU'ları Nasıl Kullanır?

RAPIDS, makine öğrenimi algoritmalarının GPU sürümlerini sağlayacak GPU hızlandırmalı makine öğrenimi kitaplıklarının bir koleksiyonudur. RAPIDS ayrıca bir veri bilimi ardışık düzenine sorunsuz bir şekilde entegre olan grafik analiz kitaplıkları içerir.

Yerel GPU bellek içi görselleştirme kitaplıkları çalışmalarda bulunmaktadır. RAPIDS, çok büyük veri kümeleriyle görselleştirme için Apache Arrow'a dayalı veri görselleştirme kitaplıkları eklemeyi planlıyor.

### RAPIDS API'nin kütüphaneleri:

#### cuDF : 

cuDF, verileri yüklemek, birleştirmek, toplamak, filtrelemek ve başka bir şekilde işlemek için bir Python GPU DataFrame kitaplığıdır ( Apache Arrow sütunlu bellek biçiminde oluşturulmuştur).

### cuML : 

cuML, diğer RAPIDS projeleriyle uyumlu API'leri paylaşan makine öğrenimi algoritmaları ve matematiksel temel işlevler uygulayan bir kitaplık paketidir.

### cuPy :

Python'da NumPy Ne ise cuPy de tam olarak odur. Sadece GPU'da çalşan versiyonu gibi düşünebilirsiniz.

### cuGraph : 

cuGraph, RAPIDS veri bilimi platformuna sorunsuz bir şekilde entegre olan NetworkX (NetworkX, grafikleri ve ağları incelemek için bir Python kütüphanesidir)  gibi işlevselliğe sahip, GPU hızlandırmalı bir grafik analiz kitaplığıdır.

### CLX : 

Cyber Log Accelerators (CLX), güvenlik analistleri, veri bilimcileri ve mühendislerin RAPIDS ve GPU hızlandırmayı gerçek dünyadaki siber güvenlik kullanım durumlarına hızla uygulamaya başlamaları için bir RAPIDS örnekleri koleksiyonu sağlar.

### cuxfilter : 

cuxfilter, farklı görselleştirme kitaplıkları ile GPU veri çerçevesi arasındaki bağlantıları çok fazla güçlük çekmeden sağlayan bir bağlayıcı kitaplığı görevi görür. Bu aynı zamanda kullanıcının farklı kitaplıklardan çizelgeleri tek bir gösterge panosunda kullanmasına izin verirken, etkileşimi de sağlar

### cuspatial : 

cuSpatial, çokgen içinde nokta, uzamsal birleştirme, koordinat sistemleri, şekil ilkelleri, mesafeler ve yörünge analizi dahil olmak üzere GIS iş akışlarını hızlandırmak için GPU hızlandırmalı bir C++/Python kitaplığıdır.


### cusignal : 

cuSignal cupy, Numba yararlanır ve ekran kartında RAPIDS ekosistem sinyali hızlandırma. Bazı durumlarda, cuSignal, CuPy aracılığıyla GPU hesaplama kaynaklarından yararlanmak için doğrudan bir Scipy Signal bağlantı noktasıdır, ancak ayrıca seçilen işlevler için ek hızlandırmalar için Numba CUDA çekirdekleri içerir.

### cuCIM : 

RAPIDS cuCIM, biyomedikal görüntülemeye odaklanan N-Boyutlu görüntüler için GPU hızlandırmalı G/Ç, bilgisayarla görme ve görüntü işleme temelleri sağlamak üzere tasarlanmış genişletilebilir bir araç takımıdır.

### kvikIO : 

kvikIO, GPUDirectStore'u (GDS) etkinleştiren cuFile'a bağlantılar sağlayan bir Python kitaplığıdır.

### RAFT :

RAFT, veri bilimi, grafik ve makine öğrenimi için yaygın olarak kullanılan temel algoritmaları ve ilkelleri içerir. Algoritmalar CUDA hızlandırmalı ve hızlı bir şekilde analitiği oluşturmak için yapı taşları oluşturuyor.

### Dask-CUDA :  
CUDA'nın etkin olduğu sistemlerde Dask çalışanlarının dağıtımını ve yönetimini iyileştirmek için çeşitli yardımcı programlar.

### RMM : 
RAPIDS Bellek Yöneticisi (RMM), cuDF (C++ ve Python) ve diğer RAPIDS kitaplıklarındaki tüm aygıt bellek ayırmaları için merkezi bir yerdir. Ayrıca, CUDA Cihaz Belleği (ve CUDA Yönetilen Belleği) için bir yedek ayırıcı ve CUDA cihaz belleği tahsisini / yeniden tahsisini daha hızlı ve eşzamansız hale getirmek için bir havuz ayırıcıdır.

## RAPIDS Kütüphaneleri:

### libcudf : 

libcudf, standart veri çerçevesi işlemlerini uygulamak için bir C/C++ CUDA kitaplığıdır.

### libcugraph : 

libcugraph, grafik işleme için bir C/C++ kütüphanesidir. Kitaplık, CUDA tabanlı grafik süreçlerine düşük düzeyde erişim isteyen kullanıcılar için bir C++ API'si sağlar.

### libcuml : 

libcuml, cuML için bir C/C++ CUDA kitaplığıdır.

Ayrıca, CUDA Cihaz Belleği (ve CUDA Yönetilen Belleği) için bir yedek ayırıcı ve CUDA cihaz belleği tahsisini / yeniden tahsisini daha hızlı ve eşzamansız hale getirmek için bir havuz ayırıcıdır.

### libkvikio :
